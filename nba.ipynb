```python
from nba_api.stats.endpoints import boxscoretraditionalv2

def simple_box_score(game_id):
    """Simple function to get and display box score."""
    
    try:
        # Get player stats
        player_stats = boxscoretraditionalv2.BoxScoreTraditionalV2(
            game_id=game_id
        ).player_stats.get_data_frame()
        
        if player_stats.empty:
            print("No data found!")
            return None
        
        # Display in a clean format
        print(f"=== BOX SCORE FOR GAME {game_id} ===")
        print(f"Total players: {len(player_stats)}")
        
        # Group by team
        for team_abbr in player_stats['TEAM_ABBREVIATION'].unique():
            team_players = player_stats[player_stats['TEAM_ABBREVIATION'] == team_abbr]
            print(f"\n{team_abbr} Players:")
            print("-" * 40)
            
            # Show key stats for each player
            for _, player in team_players.iterrows():
                print(f"{player['PLAYER_NAME']:25} {player['MIN']:>6} {player['PTS']:>3}PTS "
                      f"{player['REB']:>3}REB {player['AST']:>3}AST")
        
        return player_stats
        
    except Exception as e:
        print(f"Error: {e}")
        return None


# Use it
game_id = input("Enter NBA Game ID: ").strip()
simple_box_score(game_id)
```

```python
from nba_api.stats.endpoints import scoreboardv2
from datetime import datetime, timedelta

# Start from today
current_date = datetime.now()

# Loop through next 7 days
for i in range(100):
    date_str = current_date.strftime('%Y-%m-%d')
    
    print(f"\nChecking games for: {date_str}")
    
    try:
        scoreboard = scoreboardv2.ScoreboardV2(game_date=date_str)
        games = scoreboard.get_data_frames()[0]
        
        if len(games) > 0:
            print(f"Found {len(games)} games:")
            for _, game in games.iterrows():
                print(f"  {game['GAME_ID']} {game['GAME_STATUS_TEXT']} {game['GAMECODE']} {game['ARENA_NAME']}")
        else:
            print("No games today")
            
    except Exception as e:
        print(f"Error: {e}")
    
    # Move to next day
    current_date += timedelta(days=1)
```

```python
def test_cdn_nba_com(game_id):
    """Test cdn.nba.com endpoint used by NBA.com"""
    print("\n=== Testing cdn.nba.com live data ===")
    
    url = f"https://cdn.nba.com/static/json/liveData/boxscore/boxscore_{game_id}.json"
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
        'Accept': 'application/json, text/plain, */*',
        'Referer': f'https://www.nba.com/game/{game_id}',
        'Origin': 'https://www.nba.com',
    }
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        print(f"Status Code: {response.status_code}")
        
        if response.status_code == 200:
            data = response.json()
            print("âœ“ Success!")
            
            # Check what data we got
            if 'game' in data:
                game_info = data['game']
                print(f"Game ID: {game_info.get('gameId')}")
                print(f"Game Status: {game_info.get('gameStatusText')}")
            
            if 'stats' in data:
                stats = data['stats']
                print(f"\nStats available:")
                for key in stats.keys():
                    print(f"  - {key}")
                
                # Show active players if available
                if 'activePlayers' in stats:
                    print(f"\nActive Players: {len(stats['activePlayers'])}")
                    for player in stats['activePlayers'][:3]:  # First 3
                        print(f"  {player.get('firstName')} {player.get('familyName')}: {player.get('points')} pts")
            
            return True
        else:
            print(f"âœ— HTTP Error: {response.status_code}")
            return False
            
    except Exception as e:
        print(f"âœ— Exception: {e}")
        return False

# Run the test
test_cdn_nba_com('0022500501')
```
